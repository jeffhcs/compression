```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Path to the attributes file
attr_path = 'celeba/list_attr_celeba.txt'

# Load attributes
attributes = pd.read_csv(attr_path, delim_whitespace=True, header=1)

# Filter columns
filtered_attributes = attributes[(attributes['Attractive'] == 1) & 
                                 (attributes['Young'] == 1) & 
                                 (attributes['Blurry'] == -1)]

filtered_attributes = filtered_attributes.head(5000)

file_names = filtered_attributes.index.tolist()

file_names
```

```python
import pandas as pd
from PIL import Image
from facenet_pytorch import MTCNN
import os
from tqdm import tqdm

# Initialize MTCNN
mtcnn = MTCNN(keep_all=True)

# Path to the directory containing images
image_dir = 'celeba/img_align_celeba'

# Result list to hold bounding box data
results = []

# Iterate over each file
for file_name in tqdm(file_names):
    file_path = os.path.join(image_dir, file_name)
    image = Image.open(file_path)
    boxes, _ = mtcnn.detect(image)
    
    # Check if exactly one face is detected
    if boxes is not None and len(boxes) == 1:
        box = boxes[0]  # Get the single bounding box
        # Append filename and bounding box data to results
        results.append([file_name, *box])

# Convert results to DataFrame and save to CSV
df = pd.DataFrame(results, columns=['filename', 'x1', 'y1', 'x2', 'y2'])
df.to_csv('bounding_boxes.csv', index=False)

print("Bounding boxes saved to bounding_boxes.csv")
```
